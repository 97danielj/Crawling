# 파싱(bs4)

```python
import requests
from bs4 import BeautifulSoup
```

```python
resp = requests.get('https://naver.com')
resp
```

```python
soup  = BeautifulSoup(resp.text)
```

```python
type(soup)
```

```python
group_nav_tag= soup.find('div',class_='group_nav')
group_nav_tag
```

```python
type(group_nav_tag)
```

```python
group_nav_tag = soup.find_all('div',class_ = 'group_nav')
type(group_nav_tag)
```

```python
a_tags = group_nav_tag[0].find_all('a')
```

```python
sample = a_tags[0]
```

```python
sample.text
```

```python
sample.get('href')
```

```python
result = [{'url' :tag.get('href'), 'name' : tag.text} for tag in a_tags]
result
```

```python
data = soup.select('div#NM_FAVORITE')
data
```
